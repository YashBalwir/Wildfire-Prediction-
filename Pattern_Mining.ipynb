```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Select relevant columns
columns_to_analyze = [
    'SOURCE_SYSTEM_TYPE', 'NWCG_REPORTING_AGENCY', 'NWCG_CAUSE_CLASSIFICATION',
    'NWCG_GENERAL_CAUSE', 'STATE', 'DISCOVERY_SEASON', 'FIRE_SIZE_CLASS'
]

data_mining = data[columns_to_analyze]

# Convert categorical columns into one-hot encoding for transaction format
one_hot_encoded = pd.get_dummies(data_mining)

# Perform Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(one_hot_encoded, min_support=0.05, use_colnames=True)

# Debug: Check the structure of frequent_itemsets
print("Frequent Itemsets:")
print(frequent_itemsets.head())

# Ensure we have necessary columns
if 'support' in frequent_itemsets.columns and 'itemsets' in frequent_itemsets.columns:
    # Generate association rules from the frequent itemsets
    rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0, num_itemsets=3)
    # Sort rules by lift
    rules = rules.sort_values(by='lift', ascending=False)

    # Display rules
    print("\nAssociation Rules:")
    print(rules)
else:
    print("frequent_itemsets does not have the required structure. Check input data or apriori parameters.")
```

```python
rules.head()
```

```python
top_rules = rules.sort_values(by="lift", ascending=False).head(3)
```
